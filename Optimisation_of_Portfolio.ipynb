```python
import yfinance as yf
```

```python
TATA = yf.download(tickers='TATAMOTORS.NS', period='500d', interval='1d')
TATA=TATA.drop(['Open','High','Low','Close','Volume'], axis=1)
TATA.rename(columns = {'Adj Close':'Adj CloseT'}, inplace = True)
TATA
```

```python
REL = yf.download(tickers='RELIANCE.NS', period='500d', interval='1d')
REL=REL.drop(['Open','High','Low','Close','Volume'], axis=1)
REL.rename(columns = {'Adj Close':'Adj CloseR'}, inplace = True)
REL
```

```python
AIRT = yf.download(tickers='BHARTIARTL.NS', period='500d', interval='1d')
AIRT=AIRT.drop(['Open','High','Low','Close','Volume'], axis=1)
AIRT.rename(columns = {'Adj Close':'Adj CloseA'}, inplace = True)
AIRT
```

```python
HDFC = yf.download(tickers='HDFCBANK.NS', period='500d', interval='1d')
HDFC=HDFC.drop(['Open','High','Low','Close','Volume'], axis=1)
HDFC.rename(columns = {'Adj Close':'Adj CloseH'}, inplace = True)
HDFC
```

```python
ITC = yf.download(tickers='ITC.NS', period='500d', interval='1d')
ITC=ITC.drop(['Open','High','Low','Close','Volume'], axis=1)
ITC.rename(columns = {'Adj Close':'Adj CloseI'}, inplace = True)
ITC
```

```python
BPCL = yf.download(tickers='BPCL.NS', period='500d', interval='1d')
BPCL=BPCL.drop(['Open','High','Low','Close','Volume'], axis=1)
BPCL.rename(columns = {'Adj Close':'Adj CloseB'}, inplace = True)
BPCL
```

```python
INFOSYS = yf.download(tickers='INFY.NS', period='500d', interval='1d')
INFOSYS=INFOSYS.drop(['Open','High','Low','Close','Volume'], axis=1)
INFOSYS.rename(columns = {'Adj Close':'Adj CloseINF'}, inplace = True)
INFOSYS
```

```python
import pandas as pd
```

```python
df = pd.concat([TATA, REL,AIRT,HDFC,ITC,BPCL,INFOSYS], axis=1)
df
```

```python
df.to_csv("StockData.csv")
```

```python
import pandas as pd
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
mu = expected_returns.mean_historical_return(df)
S = risk_models.sample_cov(df)
```

```python
S
```

```python
# Optimize for maximal Sharpe ratio
ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe()
ef.portfolio_performance(verbose=True)
```

```python
cleaned_weights = ef.clean_weights()
ef.save_weights_to_file("weights.txt")  # saves to file
print(cleaned_weights)
```

```python
ef = EfficientFrontier(mu, S, weight_bounds=(-1,1))
```

```python
ef.efficient_return(target_return=0.2, market_neutral=True)
```
